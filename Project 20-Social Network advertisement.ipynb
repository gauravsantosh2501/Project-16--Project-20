Part 1 — Data Preprocessing

```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Importing the dataset
df = pd.read_csv('Social Network advertisement.csv')
df
```

```python
#printing the first few entries of the Dataset
print(df.head())
```

```python
X = df.iloc[:, [2, 3]].values
```

```python
#Storing the dependent variable in y i.e. Purchased which is 1 if user purchases the car and 0 otherwise.
y = df.iloc[:, 4].values
```

```python
#Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
#Feature Scaling#
#applying feature scaling because we want accurate prediction i.e. we want to predict which 
#users are going to buy the car or not.
#Importing Libraries
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

```python
print(X_train)
print(X_test)
```

Part 2 — Fitting our k-N.N. Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
```

```python
#Now we fit the classifier object to our training set
classifier.fit(X_train, y_train)
```

Part 3 — Predicting the Test set results

```python
y_pred = classifier.predict(X_test)
```

```python
#Displaying out the predicted values
print(y_pred)
```

```python
#Now to calculate the accuracy of our model…
```

```python
c=0
for i in range(0,len(y_pred)):
    if(y_pred[i]==y_test[i]):
        c=c+1
accuracy=c/len(y_pred)
print("Accuracy is")
print(accuracy)
```

Here Accuracy is 93%

Part-4 — Data Visualization and Confusion matrix

```python
#Visualising the Training set results
from matplotlib.colors import ListedColormap
X_set, y_set = X_train, y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('K-NN (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
